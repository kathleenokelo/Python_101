```python
## create a program that checks if two given strings are anagrams of each other.

def anagrams(string1,string2):
    string1=string1.lower().replace(' ','')
    string2=string2.lower().replace(' ','')
    
    return sorted(string1) == sorted(string2)

string1 = input('Enter your first string: ')
string2 = input('Enter your second string: ')

if anagrams(string1,string2):
    print(f'{string1} and {string2} are anagrams')
else:
    print('they are not anagrams')
```

```python
## Write a program that opens an existing text file, reads its contents,
#  and creates a new file with a different name, writing the copied content to the new file.

sourceFile = input('Enter your source file name: ')
destinationFile = input('Enter your destination file name: ')

try:
    with open(sourceFile, 'r') as sf:
        content = sf.read()

    with open(destinationFile,'w') as df:
        df.write(content)

except:
    print('An error has occured')
    quit()
```

```python
## Create a tuple to store mathematical operators (+, -, *, /) and
#  write a program that allows users to choose an operator and perform basic calculations with two numbers.

operators = ('+','-','*','/')

def operations(a,b,c):
   
    if c == '+' :
        return  a + b
    elif c == '-':
        return  a-b
    elif c == '*':
        return  a*b
    elif c == '/':
        return  a/b

    else:
        return None
    
a = int(input('Enter the first number: '))
b = int(input('Enter the second number: ' ))
c = input('Enter an operator(+,-,*,/)  ')

if c in operators:
    results = operations(a,b,c)
    print(f'the result of {a} {c} {b} is {results}')

else:
    print('Invalid operator')
```

```python
## Create a dictionary to store product information (name, price, quantity). 
# Allow users to add, update, and remove items from the inventory.

def add_product(inventory, name, price, quantity):
    inventory[name] = {'price': price, 'quantity': quantity}
    print(f"Product {name} added successfully.")

def update_product(inventory, name, price=None, quantity=None):

    if name in inventory:
        if price is not None:
            inventory[name]['price'] = price
        if quantity is not None:
            inventory[name]['quantity'] = quantity
        print(f"Product {name} updated successfully.")
    else:
        print(f"Product {name} not found in inventory.")

def remove_product(inventory, name):
    
    if name in inventory:
        del inventory[name]
        print(f"Product {name} removed successfully.")
    else:
        print(f"Product {name} not found in inventory.")

def display_inventory(inventory):
    
    if not inventory:
        print("Inventory is empty.")
    else:
        print("Current Inventory:")
        for name, details in inventory.items():
            print(f"Name: {name}, Price: {details['price']}, Quantity: {details['quantity']}")

def main():
    inventory = {}
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Update Product")
        print("3. Remove Product")
        print("4. Display Inventory")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter product name: ")
            price = float(input("Enter product price: "))
            quantity = int(input("Enter product quantity: "))
            add_product(inventory, name, price, quantity)
        elif choice == '2':
            name = input("Enter product name: ")
            price = input("Enter new product price (leave blank to keep current): ")
            quantity = input("Enter new product quantity (leave blank to keep current): ")
            price = float(price) if price else None
            quantity = int(quantity) if quantity else None
            update_product(inventory, name, price, quantity)
        elif choice == '3':
            name = input("Enter product name to remove: ")
            remove_product(inventory, name)
        elif choice == '4':
            display_inventory(inventory)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
## Create a dictionary where keys are departments in a company and
#  values are nested dictionaries containing employee information (name, role).

def add_employee(departments, department, name, role):
    
    if department not in departments:
        departments[department] = {}
    departments[department][name] = role
    print(f"Employee {name} added to {department} department successfully.")

def update_employee(departments, department, name, role):
   
    if department in departments and name in departments[department]:
        departments[department][name] = role
        print(f"Employee {name}'s role updated to {role} in {department} department.")
    else:
        print(f"Employee {name} not found in {department} department.")

def remove_employee(departments, department, name):
    
    if department in departments and name in departments[department]:
        del departments[department][name]
        print(f"Employee {name} removed from {department} department.")
    else:
        print(f"Employee {name} not found in {department} department.")

def display_departments(departments):
  
    if not departments:
        print("No departments available.")
    else:
        print("Departments and Employees:")
        for department, employees in departments.items():
            print(f"Department: {department}")
            for name, role in employees.items():
                print(f"  Name: {name}, Role: {role}")

def main():
    departments = {}
    while True:
        print("\nDepartment Management System")
        print("1. Add Employee")
        print("2. Update Employee")
        print("3. Remove Employee")
        print("4. Display Departments")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            department = input("Enter department name: ")
            name = input("Enter employee name: ")
            role = input("Enter employee role: ")
            add_employee(departments, department, name, role)
        elif choice == '2':
            department = input("Enter department name: ")
            name = input("Enter employee name: ")
            role = input("Enter new employee role: ")
            update_employee(departments, department, name, role)
        elif choice == '3':
            department = input("Enter department name: ")
            name = input("Enter employee name: ")
            remove_employee(departments, department, name)
        elif choice == '4':
            display_departments(departments)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()


```

```python
## Write a function to create a new set that contains the identical items from two given sets.

#Return the identical elements in the two given sets.
#For inputs with sets {1, 2, 3, 4} and {3, 4, 5, 6}, the return value should be either {3, 4} or {4, 3}.

def find_identical_elements(set1, set2):
    return set1.intersection(set2)

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
identical_elements = find_identical_elements(set1, set2)
print(f"The identical elements in the two given sets are: {identical_elements}")
```

```python
## Write a Python program to check that a string 
# contains only a certain set of characters (in this case a-z, A-Z and 0-9).
import re

def allowed_chars(input_string):
   
    pattern = "^[a-zA-Z0-9]*$"
    return bool(re.match(pattern, input_string))


test_string = input("Enter your string: ")
if allowed_chars(test_string):
    print("The string contains only a-z, A-Z, and 0-9 characters.")
else:
    print("The string contains other characters besides a-z, A-Z, and 0-9.")

```

```python
## Write a program that uses regex to validate user input. Examples:
#Check if a password meets certain criteria (e.g., minimum length, containing uppercase and lowercase letters).
# validate a date format (e.g., DD/MM/YYYY).


import re

def validate_password(password):
    if len(password) < 8:
        return False
    
    if not re.search(r'[A-Z]', password):
        return False
    
    if not re.search(r'[a-z]', password):
        return False
    
    if not re.search(r'[0-9]', password):
        return False
    
    if not re.search(r'[@$!%*?&#]', password):
        return False
    
    return True

def validate_date(date):
    pattern = r'^\d{2}/\d{2}/\d{4}$'
    return bool(re.match(pattern, date))


passwords = input('Enter your password: ')
dates = input('Enter your date: ')

print("Password Validation:")
if validate_password(passwords):
    print(f"'{passwords}': {validate_password(passwords)}")
else:
    print('Password is not valid')

print("Date Validation:")
if validate_date(dates):
    print(f"'{dates}': {validate_date(dates)}")
else:
    print('Date is not valid')

```
